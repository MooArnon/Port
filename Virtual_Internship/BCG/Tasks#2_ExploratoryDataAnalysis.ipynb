## Data Visualization

### Create Function for plotting the graph

This task will use 
1. Distribution: for visualizing ratio of churn and stay for each features.
2. Box plot: for visualizing distribution of data.
3. Percentage bar plot: for visualizing churning ratio of of each condition in feature
4. Correlation heatmap: for visualizing correlated between features.

```python
def distribution_plot(df, feature, size=(16,6), bins=50):
   # Create dataframe that will be used for plotting
   df = pd.DataFrame({"Stay":client_data[client_data['churn'] == 0][feature],
                     "Churn": client_data[client_data['churn'] == 1][feature]})
   # Plot data
   df[['Stay', 'Churn']].plot(kind='hist',stacked=True, figsize=size, bins=bins)
   plt.xlabel(feature)
   plt.show()

def multiple_box_plot(df, features):
    # Create plot frame
    fig, axs = plt.subplots(nrows=len(features), figsize=(18,20))
    x=0 # time reference
    for i in features:
        # Plot data
        sns.boxplot(data=df, x=i, hue='churn', ax=axs[x])
        x+=1
    plt.show()

def percentage_bar_plot(df, feature, rotation_=0):
    df = df.groupby([feature, 'churn'])['id'].count().unstack(level=1)
    final_total = df.div(df.sum(axis=1), axis=0)*100
    ax = final_total.plot(kind='bar', stacked=True, figsize=(10,7))
    plt.xticks(rotation=rotation_)
    # Annotate
    for p in ax.patches:
        # Bar
        width = p.get_width()
        height = round(p.get_height(),1)
        # Position
        x, y = p.get_xy()
        # Pass if churn = 0
        if height == 0.0:
                continue
        # Code for adding annotate
        ax.annotate(height, ((x-0.15+width/2, y+height*1.01/2)))
    plt.legend(["Stay", "Churn"], loc='lower right')   # 0=stay, 1=churn
```

## Hypothesis checking

In order to analyze hypothesis, Price sensitivity, we need to divide data in to small section. This method give us a precise result.

```python
# Finding mean of each time frame
## DF categorized by date time
date_3m = price_data[price_data['price_date'] >= '2015-10-01']
date_6m = price_data[price_data['price_date'] >= '2015-06-01']
## DF of mean of value by date time
mean_date_3m = date_3m.groupby('id').mean().reset_index()
mean_date_6m = date_6m.groupby('id').mean().reset_index()
mean_date_year = price_data.groupby('id').mean().reset_index()

# Change columns name of each data frame
## 3m
col_3m = {"price_off_peak_var" : "mean3m_price_p1_var",
          "price_peak_var"     : "mean3m_price_p2_var",
          "price_mid_peak_var" : "mean3m_price_p3_var",
          "price_off_peak_fix" : "mean3m_price_p1_fix",
          "price_peak_fix"     : "mean3m_price_p2_fix",
          "price_mid_peak_fix" : "mean3m_price_p3_fix"
}
mean_date_3m = mean_date_3m.rename(columns=col_3m)
mean_date_3m['mean3m_price_p1'] = mean_date_3m['mean3m_price_p1_var'] + mean_date_3m['mean3m_price_p1_fix']
mean_date_3m['mean3m_price_p2'] = mean_date_3m['mean3m_price_p2_var'] + mean_date_3m['mean3m_price_p2_fix']
mean_date_3m['mean3m_price_p3'] = mean_date_3m['mean3m_price_p3_var'] + mean_date_3m['mean3m_price_p3_fix']

## 6m
col_6m = {"price_off_peak_var" : "mean6m_price_p1_var",
          "price_peak_var"     : "mean6m_price_p2_var",
          "price_mid_peak_var" : "mean6m_price_p3_var",
          "price_off_peak_fix" : "mean6m_price_p1_fix",
          "price_peak_fix"     : "mean6m_price_p2_fix",
          "price_mid_peak_fix" : "mean6m_price_p3_fix"
}
mean_date_6m = mean_date_6m.rename(columns=col_6m)
mean_date_6m['mean6m_price_p1'] = mean_date_6m['mean6m_price_p1_var'] + mean_date_6m['mean6m_price_p1_fix']
mean_date_6m['mean6m_price_p2'] = mean_date_6m['mean6m_price_p2_var'] + mean_date_6m['mean6m_price_p2_fix']
mean_date_6m['mean6m_price_p3'] = mean_date_6m['mean6m_price_p3_var'] + mean_date_6m['mean6m_price_p3_fix']

## year
col_12m = {"price_off_peak_var" : "mean12m_price_p1_var",
          "price_peak_var"     : "mean12m_price_p2_var",
          "price_mid_peak_var" : "mean12m_price_p3_var",
          "price_off_peak_fix" : "mean12m_price_p1_fix",
          "price_peak_fix"     : "mean12m_price_p2_fix",
          "price_mid_peak_fix" : "mean12m_price_p3_fix"
}
mean_date_12m = mean_date_year.rename(columns=col_12m)
mean_date_12m['mean12m_price_p1'] = mean_date_12m['mean12m_price_p1_var'] + mean_date_12m['mean12m_price_p1_fix']
mean_date_12m['mean12m_price_p2'] = mean_date_12m['mean12m_price_p2_var'] + mean_date_12m['mean12m_price_p2_fix']
mean_date_12m['mean12m_price_p3'] = mean_date_12m['mean12m_price_p3_var'] + mean_date_12m['mean12m_price_p3_fix']
```

```python
# Merge data
price_data_final = pd.merge(mean_date_12m, mean_date_6m, on='id')
price_data_final = pd.merge(price_data_final, mean_date_3m, on='id')
## For investigating hypothesis
price_churn = pd.merge(price_data_final, client_data[['id', 'churn']], on='id')

price_data_final['price_date']
```
