## Data preparation

### transaction_data

#### DATE

```python
print(transaction_data.shape)
transaction_data.info()
```

```python
import datetime as dt
# Convert time stamp to datetime
transaction_data['DATE'] = pd.TimedeltaIndex(transaction_data['DATE'], unit='d')+ dt.datetime(1899,12,30)
# Check that there are not future date
today_date = dt.date.today()
assert transaction_data['DATE'].max() <= today_date 
# Check type of feature
assert transaction_data['DATE'].dtype == 'datetime64[ns]' 
# Sort data frame by date
transaction_data = transaction_data.sort_values(by='DATE')
```

```python
# Missing date ?
date_check = transaction_data.set_index('DATE')
date_check.index = pd.to_datetime(date_check.index)
pd.date_range(start='2018-07-02', end='2019-06-30').difference(date_check.index)# To see missing date.
```

```python
# Plot missing data
sns.set(rc={'figure.figsize':(25,12)})
date_plot = transaction_data[(transaction_data['DATE'] > '2018-12-01') & (transaction_data['DATE'] <= '2019-01-15')]
sns.countplot(data=date_plot, x='DATE')
plt.xticks(rotation=90)
plt.show()
```

From data plotted, number of transactions is dramatically increase before the missing date. This in sight show us the normal behavior of customer. 25th December is christmas day the store may close or people are stay with their family. Therefore the days before christmas day will have very high customer compare with the others.The missing date, `2018-12-25`, data can be ignored.

#### PROD_NAME

We can extract some interesting features, BRAND_NAME and BAG_SIZE.

```python
# BRAND_NAME
bag_brands = pd.DataFrame(transaction_data.PROD_NAME.str.split().str.get(0)) # Split from space
transaction_data['BRAND_NAME'] = bag_brands
transaction_data['BRAND_NAME'].value_counts().reset_index().sort_values(by='index')
```

Some brand using more than 2 names. for instant, `Smith` and `Smiths`. This problem need to be fix be for doing next features.

```python
## Clean BRAND_NAME
### im not sure in this section. Some brand was only in USA or EU. I can not figure out some mistaken.
brand_name_dict = {'Smith':'Smiths', 'WW':'Woolworths', 'Snbts':'Sunbites', 'RRD':'RED',
                    'Infzns':'Infuzions', 'Dorito':'Doritos', 'GrnWves':'Grain'}
transaction_data['BRAND_NAME'] = transaction_data['BRAND_NAME'].replace(brand_name_dict)
### Check
for i in transaction_data['BRAND_NAME']:
    assert i != ['Smith', 'WW', 'Snbts', 'Red', 'Infzns', 'Dorito', 'GrnWves']
```

#### BAG_SIZE

```python
# BAG_SIZE
import re
# Fn to extract number from str
def find_number(text):
    num= re.findall(r'[0-9]+',text)
    return " ".join(num)
# Extract
transaction_data['BAG_SIZE'] = transaction_data['PROD_NAME'].apply(lambda x: find_number(x))
# Convert to int
transaction_data['BAG_SIZE'] = transaction_data['BAG_SIZE'].astype(int)
# Check outliers
transaction_data['BAG_SIZE'].sort_values(ascending=False).head(4)
```

## Data Analysis

### LIFESTAGE and PREMIUM_CUSTOMER
To see market share ratio from customer type.

```python
# Build dataframe
Lifestage_PremiumCustomer_df = final_data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER']).agg({'LYLTY_CARD_NBR':'count'}).reset_index()
# Total
total = sum(Lifestage_PremiumCustomer_df['LYLTY_CARD_NBR'])
# Calculate market share 
Lifestage_PremiumCustomer_df['market_share'] = round((Lifestage_PremiumCustomer_df['LYLTY_CARD_NBR']/total)*100, 2)
Lifestage_PremiumCustomer_df['market_share_str'] = Lifestage_PremiumCustomer_df['market_share'].astype(str)
# Build Label
Lifestage_PremiumCustomer_df['label'] = Lifestage_PremiumCustomer_df['LIFESTAGE'] + " | " + Lifestage_PremiumCustomer_df['PREMIUM_CUSTOMER'] + '\n' + Lifestage_PremiumCustomer_df['market_share_str']
Lifestage_PremiumCustomer_df
```

```python
# Build tree plot
import squarify
import matplotlib
cmap = matplotlib.cm.Blues
mini=min(Lifestage_PremiumCustomer_df['market_share'])
maxi=max(Lifestage_PremiumCustomer_df['market_share'])
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in Lifestage_PremiumCustomer_df['market_share']]
# Change color
squarify.plot(sizes=Lifestage_PremiumCustomer_df['market_share'],label=Lifestage_PremiumCustomer_df['label'] ,alpha=.8, color=colors )
plt.axis('off')
plt.show()
```

`OLDER FAMILIES | Budget`, `RETIREES | Mainstream`, `YOUNG SINGLES/COUPLES | Mainstream` and `YOUNG FAMILIES | Budget` are the TOP4 of highest transaction, high sorted respectively. Which have market share more than 7% of overall. We can focus on these set of customer to analyze buying behavior.

```python
final_data['LifeStage_Premium'] = final_data['LIFESTAGE'] + '_' +final_data['PREMIUM_CUSTOMER']
focus = ['OLDER FAMILIES_Budget', 'RETIREES_Mainstream', 'YOUNG SINGLES/COUPLES_Mainstream', 'YOUNG FAMILIES_Budget']
```

### Bag's size

```python
for i in focus:
    df = final_data[final_data['LifeStage_Premium'] == i]
    df = df.groupby('BAG_SIZE').agg({'TXN_ID':'count'}).reset_index()
    df.plot.bar(x='BAG_SIZE', y='TXN_ID', figsize=(11, 5))
    plt.title(i)
    
```

This section shows us that 175, 150, 134 and 110 are the highest sales.

### Brand

```python
for i in focus:
    df = final_data[final_data['LifeStage_Premium'] == i]
    df = df.groupby('BRAND_NAME').agg({'TXN_ID':'count'}).reset_index()
    df.plot.bar(x='BRAND_NAME', y='TXN_ID', figsize=(11, 5))
    plt.title(i)
```

Kattle, Smiths and Doritos is the highest sales

### Quality of buying

```python
for i in focus:
    df = final_data[final_data['LifeStage_Premium'] == i]
    df = df.groupby('PROD_QTY').agg({'TXN_ID':'count'}).reset_index()
    df.plot.bar(x='PROD_QTY', y='TXN_ID', figsize=(11, 5))
    plt.title(i)
```

Customers are most likely buy a couple pack of ships.
