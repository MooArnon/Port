## Strategy performance measurement

This task will be combine all metric to be 1 months for reduce latency and increase accuracy of finding control store

```python
# combine date in to one month
data['year_month'] = data['DATE'].dt.to_period('M')
data.head(4)
```

We need to drop the data of store that less than 12 months for precision of the verification. 

```python
# Drop store that has history less than 12 months.
store_less_than_12m = data.groupby('STORE_NBR')['year_month'].nunique()
store_less_than_12m = store_less_than_12m[store_less_than_12m != 12]
store_less_than_12m = store_less_than_12m.index.to_list()
print('Store which less than 12m: ', store_less_than_12m)
dropping_data = data[data['STORE_NBR'].isin(store_less_than_12m)].index
data = data.drop(dropping_data)
```

### Metrics

```python
# Create metric features
data['year_sale'] = data.groupby('STORE_NBR')['TOT_SALES'].transform('sum')
data['year_cust'] = data.groupby('STORE_NBR')['LYLTY_CARD_NBR'].transform('nunique')
data['month_sale'] = data.groupby(['STORE_NBR', 'year_month'])['TOT_SALES'].transform('sum')
data['month_cust'] = data.groupby(['STORE_NBR', 'year_month'])['LYLTY_CARD_NBR'].transform('nunique')
```

```python
avg_trans = data.groupby('STORE_NBR').apply(lambda subdf: (subdf['TXN_ID'].nunique() / subdf['year_cust'].unique()))
avg_trans = avg_trans[avg_trans.str.len()>0]
avg_trans = avg_trans.astype('float64')
data['avg_txn_per_cust'] = data['STORE_NBR'].map(avg_trans)
data
```

```python
# Sepperate before and after trial period 
pre_df = data[data['DATE'] < "2019-02-01"] # before
trial_df = data[(data['DATE'] > "2019-01-31") & (data['DATE'] < "2019-05-01")] # after
```

```python
# Finding correlation between each features by using heatmap.
corr_df = pre_df.corr() # Construct the correlation dataframe
mask = np.triu(np.ones_like(corr_df, dtype=bool)) # This line came from stack overflow. It can pass this argument, but it will return normal heatmap, which is so eyestrain.
sns.heatmap(corr_df, mask = mask, cmap = 'YlGnBu', annot = True) # Build heatmap graph
plt.xticks(rotation = 45)
plt.show()
```

```python
# Determine the metric features
metrics_cols = ['STORE_NBR', 'year_month', 'year_sale',
                'year_cust','month_sale', 'month_cust', 'avg_txn_per_cust']
```

### Calculate correlation to find controlled store

#### Function

```python
def extract_metrics(df): 
    sub_df = df.loc[:, metrics_cols].set_index(['STORE_NBR', 'year_month']).sort_values(by = ['STORE_NBR', 'year_month'])
    sub_df.drop_duplicates(inplace = True, keep = 'first')
    return sub_df 
```

```python
metrics_df = extract_metrics(pre_df)
```

```python
def cal_corr(trial_store):
    a=[]
    metrics = metrics_df[['month_sale', 'month_cust']] 
    for i in metrics.index:
        a.append(metrics.loc[trial_store].corrwith(metrics.loc[i[0]]))
    sub_df = pd.DataFrame(a)
    sub_df.index = metrics.index
    sub_df = sub_df.drop_duplicates()
    sub_df.index = [s[0] for s in sub_df.index]
    sub_df.index.name ="store_nbr"
    sub_df = sub_df.abs()
    sub_df['mean'] = sub_df.mean(axis=1)
    sub_df.sort_values(by = 'mean', ascending = False, inplace = True)
    return sub_df
```

```python
print(cal_corr(77).drop(77).head(3))
print(cal_corr(86).drop(86).head(3))
print(cal_corr(88).drop(88).head(3))
```

Trial and control stores are occur. Which can be used for analyze efficiency of new strategy. <br>

### Model Evaluate
using line plot to see how accurate of the model

```python
data['STORE_NBR'] = data['STORE_NBR'].astype(object)
eva_data = data.set_index('STORE_NBR')
data_77_233 = eva_data.loc[[77,233]].reset_index()
data_86_23 = eva_data.loc[[86,155]].reset_index()
data_88_178 = eva_data.loc[[88,178]].reset_index()
data_trial_control = [data_77_233, data_86_23, data_88_178]
```

#### Visualization

```python
# Total sales
for i in data_trial_control:
    data_plot = i.groupby(['year_month', 'STORE_NBR']).agg({'TOT_SALES' : 'sum'}).reset_index()
    sns.catplot(data=data_plot, x='year_month', hue='STORE_NBR', y='TOT_SALES', kind='point' )
    plt.xticks(rotation=90)
    plt.show()
```

Before-trial period, trial stores and controlled stores have the almost same behaviour in total sales features. Especially for store number #88, the closed behaviour of this store is store number 178. Which is quite the same graph pattern but a difference in values. These details are quite acceptable, 2 out of three have very close in behaviour, values and only behaviour for the remained. <br><br>
After-trial period, the total sale of trial stores are dramatically increased. However, at the end of the trial period, values of sales are intersect with controlled stores. For store 88, the observer can indicate almost the same pattern with other trial-control stores relationships. 

```python
# Total sales
for i in data_trial_control:
    data_plot = i.groupby(['year_month', 'STORE_NBR']).agg({'TXN_ID' : 'count'}).reset_index()
    sns.catplot(data=data_plot, x='year_month', hue='STORE_NBR', y='TXN_ID', kind='point' )
    plt.xticks(rotation=90)
    plt.show()
```

The number of customers shows the same result with total sales, trial stores do better in a trial period. <br><br>
In conclusion, from features, data visualization and analysis, the strategy has very high efficiency for increasing total sales and transactions per customer. The interested values are dramatically increased in started period, but intersect with controlled value at the end of trial period. This insight means new strategic model can impel total sales and transaction per customer.
